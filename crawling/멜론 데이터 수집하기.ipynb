# 노래제목 수집

```python
res = req.get("https://www.melon.com/chart/index.htm",headers = user_agent )
#그냥 이 url을 함수로 담아도 된다
soup = bs(res.text, "lxml")

data = soup.select('div.wrap>div.wrap_song_info>div.ellipsis.rank01>span>a')
#혹시 클래스두개 쓰고 싶으면 첫클래스.두번쨰 클래스 적기
```

```python
data
```

```python
songs_list = []
for i in data:
    print(i.text)
    songs_list.append(i.text)
songs_list
```

```python
#가수
singer = soup.select('div.wrap>div.wrap_song_info>div.ellipsis.rank02>a')
list1=[]
for i in singer:
    print(i.text)
    list1.append(i)
len(list1)
    
```

```python
#106개인 이유 이거 콜라보 해서 , 으로 이어진것을 두개로 인식해서 
#태그를 조금 덜 쓰면 됨
#a때고 div까지만  a로 가게 되면 
singers = soup.select('div.wrap>div.wrap_song_info>div.ellipsis.rank02>span.checkEllipsis')
singers
singers_list= []
for i in singers:
    print(i.text)#span태그가 가지고 있는 내용 전부 다 출력 근데 태그정보는 뺴고 contanct만
    singers_list.append(i.text)
```

```python
#지금 모은거 그냥 났두면 다 사라진다 가수와 노래정보를 csv파일로 제작
songs_list
singers_list
#제작할때는 판다스 라이브러리 안의 데이터 프레임 쓰는게 편하다
import pandas as pd
dic = {'가수':singers_list, '노래제목':songs_list}
melon  = pd.DataFrame(dic)
melon
```

```python
melon.to_csv('melon1_top100.csv',encoding= 'euc-kr')
```
