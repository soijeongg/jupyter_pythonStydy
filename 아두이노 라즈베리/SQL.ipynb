```python
import pymysql as ps
# DB와 연결하는 통로!! 
conn = ps.Connect(host='localhost',
                  db='iot',
                  user='root',
                  password='1234')

# SQL문이 전달되는 통로!!
curs = conn.cursor()

sql = 'insert into sensordb(sensing) values (1000)'
curs.execute(sql)

conn.commit()

# 열었던 역순으로 닫아주자!!
curs.close()
conn.close()
```
