```python
#flask파일 안에 있는 Flask라는 클래스
from flask import Flask, request

#플라스크 클래스를 생성하겠다
#생성된 클래스는 인스턴스라고 부름
app = Flask(__name__)
#led가 꺼져 있는 것이 기본값
result = '{"led":"0"}'
#여기사 만든 인스턴스를 찾아 루트 찾아준다 
@app.route('/')
def  index():
    return '<h1>정소이의 홈페이지</h1>'

@app.route('/state')
def state():
    global result
    result = '{"led":"1"}'
#이 global 다음에 쓰는 함수는 전역변수를 가져다 쓰겠다 라는 의미
@app.route('/led/on')
#ledon이라는 거 뜨면서 1로 변함
def led_on():
    global result
    result = '{"led":"1"}'
    return "LED ON"


@app.route('/led/off')
def led_off():
    global result
    result = "{'led':'1'}"
    return "led off"
@app.route("/hello",methods = ['GET'])
def hi():
#요청 객체 안에 담겨 있는 정보를 꺼내자
#요청 객체 request get 이라면 args["키값"] 이 apple이 약속된 키 
    
    name = request.args['apple']
    result = f"{name}님 환영합니다"
    #ahwllo라는 사이트에 이름을 보내자
    return result


if __name__ == '__main__':
    app.run(port = 5017)
    
#name이 main으로 바뀌면 실행
#run실행하면 실행 run을 만나면 break처럼 진행X


```
