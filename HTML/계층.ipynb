# 계층선택자

```python
자식 선택자:
    내가 코드를 트리 구조로 쭉 뻗었을때 바로 아래 있는 것 > 로 표현
    ul바로 밑의 li 
    ul>li{}
    이럴떄 는 오직 ul 바로 아래 있는 li만 가능
    
자손 선택자: 
    자식이자 자손이 될 수 있음
    기호 없음 띄어쓰기로 표현
    바로 아래있지 않더라도 위에 거슬러 올라가면 같은 사람(자식 선택자도 포함)
    p태그를 기준으로 아래있는 모든 span 태그들
    p span {}
    
형제관계(후행):
동등한 위치에 존재
근접후행: 바로 붙어있는 형제 관계 +
후행: 떨어져있는 형제 ~ 사용 
나와 같은 부모를 가지고 있는 요소들중 내 뒤에 있는 +
li를 기준으로 형제 li들 
li~li{}
근접후행: 
    후행관계들중 바로 내뒤에 있는 요소 li를 기준으로 바로 옆에 있는 span선택
    li+span{}
    
```
