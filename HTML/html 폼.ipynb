```python
form태그 
        웹페이지에 입력폼을 만들때 사용하는 태그 
        사용자에게 정보르 받아주는 정보 주머니 역할
        - 함수 속성 2가지:method, action
        method: 전달방식 어떻게 넘겨줄껀지?
                보안이 중요!!(post) 괜찮음(get)
                보안 중요X(예를들어 검색창)

        action : 데이터를 전달할 위치
            - 보통은 백앤드 주소
            - #을 쓰면 현재페이지 라는 뜻 공란으로 남겨주면 뭔가 그러니까
            
```

```python
  <form method="get" action ="#"> 
        <h3>입력</h3>
        <!-- input태그: 사용자가 입력할 수 있는 공간
                        정해줘야 할 속성: type =""뭘 입력할지-->
    ID: <input type ="text"></input><br>
    pW: <input type ="password"></input><br>
    강의실 :<input type ="text"></input><br>
</form>
```

```python
input태그의 속성
                (1) name: 입력한 값을 구분하기 위해 이름 지정
                (2) placeholder: 미리보기 텍스트 (회색글씨 내가 실제 입력시 사라짐)
                이거 약간 파이썬 인풋할때 인풋안에 쓰는 텍스트 처럼
                (3) autofocus: 자동 커서 실행 -->
                (4) value: 기본 값을 설정 할 수 있다
                (5) readoonly: 읽기 전용만 (수정하지 못하게 )
                (6) maxlength: 글자 수 제한 지정
```
