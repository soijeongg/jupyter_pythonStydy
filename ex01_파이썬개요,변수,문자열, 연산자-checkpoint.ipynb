#현재 가장 많이 쓰이는 방법 format 함수를 활용한 표맷팅



```python
MONTH = 6
day = 9
s7 = "오늘을 {}월 {}일 입니다.".format(MONTH, day)
print(s7)
```

```python
# f-string기법
month =6 
day =9
s8 = f"오늘을 {month}월 {day}일 입니다!"
print(s8)
```

```python
x=100
y=200
add=x+y
#%기호 사용 포매팅
print("%d와 %d의 합은 %d입니다" %(x, y, add))
# format 함수 사용 포매팅
print("{}와 {}의 합은 {}입니다" .format(x, y, add))
# f-string 사용
print(f"{x}와 {y}의 합은 {add}입니다")
```

### 논리연산자
   - 논리형(True, False)을 연산해주는 연산자
     - not = 논리를 부정해주는 연산자 논리 앞에 써주기!(!)
     - and = 둘다 True 면 True 아니면 False(&)
     - or = 둘 중 하나만 True라면 True를 출력(|)
     

```python
a=10
b=7
print(a!=b & a==b)
print(a!=b | a==b)
print(a>b)
```

### 삼항연산자
    - a if 조건식 else b
    - 조건식에는 True or False 를 가지는 식이 들어가야 하고 T면 a 실행
    - F면 b 실행
    - True는 앞을 False는 뒤를 출력한다
    - 주의
        - 조건식의 결과는 항상 boolean으로 나와야 함
   

```python
#나이가 20살 이상이면 성인 출력 20살 미만이라면 미성년자 출력
age = 18
"성인" if age >=20 else "미성년자"
```
