```python
#머신러닝  데이터에서 규칙을 찾는것
#우리가 하는게 아니라 컴퓨터한테 시켜서 
#미스터리사인이 예시
#주어진 수를 보고 규칙을 찾는 문제
```

```python
#규칙: 큰수 + 두수의차
#22 ?7 =37
```

```python
#데이터를 준비
```

```python
list1 =[]
for i in range(100):
    for j in range(100):
        result = (i+(i-j))if i>j else(j+(j-i))
        list1.append([i,j,result])
```

```python
#가운데 조건식이 있고 이 조건식이 참이면 왼쪽게  실행되고 거짓이면 오른쪽게 실행이 된다 
```

```python
#데이터 준비2
#머신러닝에 학습 할 수 있는 형태로 바꾸기
#np.array, Pd.series Pd.dataframe
import pandas as pd
data1 = pd.DataFrame(list1, columns = ['N1','N2','Result'])
data1.head()
```

```python
#데이터 준비의 3단계
#문제와 정답으로 분리하기
```

```python
X =data1.loc[:,"N1":"N2"]
y =data1.loc[:,"Result"]
```

```python
#머신러닝 모델 준비
from sklearn.ensemble import RandomForestRegressor
random_forest = RandomForestRegressor()
```

```python
모델을 번수에다가 담아준것
```

```python
random_forest.fit(X,y)
```

```python
#이게 어떤 규칙이 있는지 확인하고 있음 
#끝나게 되면 데이터에서 살펴본 가장 정확하다고 생각하는 규칙 하나를 저장
```

```python
random_forest.predict([[22,7]])
```

```python
list2 =[]
for i in range(1,100):
    for j in range(1,100):
        result = (i%j)if i>j else(j%i)
        list2.append([i,j,result])
```

```python
import pandas as pd
data1 = pd.DataFrame(list2, columns = ['N1','N2','Result'])
data1.head()
```

```python
X =data1.loc[:,"N1":"N2"]
y =data1.loc[:,"Result"]
```

```python
from sklearn.ensemble import RandomForestRegressor
random_forest = RandomForestRegressor()
```

```python
random_forest.fit(X,y)
```

```python
random_forest.predict([[19,4]])
```
