# 5. 모델선택 및 하이퍼 파라미터 튜닝

```python
#데이터 분리
#문제(X- iris['data'])와 정답(y - iris['target'])
#훈련과 평가
X =iris['data']
y =iris['target']
#훈련(train -7)과 평가 (test-3) + 데이터 섞어주기
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =train_test_split(X,y, test_size =0.3,
                                                  random_state =6)
```

```python
#이 티스트 사이즈는 테스트가 이 0.3 가져감
#랜덩은 굳이 안써도 되지만 이걸 쓰면 같은 방법을 공유하개 됨
```

```python
#데이터 분리 확인
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#KNN모델 불러오기
from sklearn.neighbors import KNeighborsClassifier
knn =KNeighborsClassifier()
```

# 6. 학습

```python
knn.fit(X_train,y_train)
```

```python
knn.score(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
knn.predict([[4.8, 3. , 1.4, 0.1]])
```

# 8. 하이퍼 파라미터 튜닝

- 5단계 모델선택및 하이퍼 파라미터 튜닝
- 6단계 : 학습
- 7단계 : 평가
- 이걸 반복하면서 최적의 값을 찾아야 함

```python
#점수를 저장할 라스트 생성
train_list =[]
test_list=[]
#n_neighbors: 이웃의 수 결정
n_n_settings =range(1,21)
for n in n_n_settings:
    #이웃의 수 조절하면서 모델 생성하기!(5단계)
    knn =KNeighborsClassifier(n_neighbors= n)
    #모델학습: 6단계
    knn.fit(X_train,y_train)
    
    #평가 7단계 지금은 어떤 하이퍼파라미터가 최적인지 보기위해 저장
    train_score = knn.score(X_train,y_train)
    test_score =knn.score(X_test,y_test )
    #평가결과 저장
    train_list.append(train_score)
    test_list.append(test_score)
```

```python
#그래프로 확인하기
import matplotlib.pyplot as plt
plt.plot(n_n_settings,train_list, label ='Train',marker ="o") 
plt.plot(n_n_settings,test_list, label ='Test',marker ="o")
plt.xlabel("n_neighbors")
plt.ylabel('accuracy')# 꺽은선 그래프
plt.grid()

plt.legend()
plt.show()
```

```python
#이웃의 수가 5일떄 가장 적절하다
```
