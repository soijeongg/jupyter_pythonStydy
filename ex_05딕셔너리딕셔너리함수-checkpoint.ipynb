## 키값이 명확하게 존재하고 value를 한쌍으로 가짐
    키값은 불변 중복이 안됨
    벨류는 중복이 가능 
    순서가 없는 집합 
    딕셔너리의 벨류값은 항상 key값이 쌍으로 연결되어 있음

```python
# 딕셔너리명 = {key:value,key:value}
# key는 변하지 않는 값을 사용 value는 변하지 않는 값 변하는 값 둘다 가능
#순서중요  X 그 key값에 어떤 value가 들어있는지가 중요
#calue에는 어떤타입이 들어가든지 상관 없음
# 이건 메소드로 사용할때 무조건 글자 뒤에 s 붙이기
```

```python
a={}
b={"name":"정소이","pn":"010-1111-2222","age":22}
c={1:"232",2:"45"}
print(type(a))
print(c)
print(b)
```

```python
#딕셔너리 값 추가하기
dict1 = {"name":"정소이","pn":"010-1111-2222","age":22}
#딕셔너리명[key] = value
dict1["birth"] ="0127"
dict1
```

```python
# value 바꾸기 있는거면 바뀌고 없는 거면 추가
dict1["pn"] = "010-3333-4444"
dict1
```

```python
#인덱싱
print(dict1["name"])
#함수를 이용해서 내가 원하는 정보를 가져오고 ㅇ싶ㅇ르때l
#dict1.get(키값) 앞에서 그냥 인덱싱 처럼 가져오는건 없으면 에러지만 get은 none
# 없는데 찾으면 none 
```

```python
if dict1.get"email"== None:
    print("없는 데이타")
```

```python
#반환 값 말고는 크게 차이x
#key낭 사디고오기 딕샤너리.keys()
keyli =list(dict1.keys())
print(keyli[0])

keyli[0]
#굳이 리스트로 변환하지 않아도 반복문으로 사용가능
valuess = list(dict1.values())
print(valuess[0])
for value in dict1.values():
        print(value)
```
