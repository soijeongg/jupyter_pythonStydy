```python

def divisor(num):
    num_list = []
    for i in range(1,num+1):
        if num %i==0:
            num_list.append(i)
    return num_list
print(divisor(40))
```

```python
def cal(num1, num2,op):
    """dlfjgp """
    if op == "+":
        result= num1+num2
    elif op == "-":
        result = num1-num2
    return result
num1 = int(input("첫번째 정수 입력 >>"))
num2 = int(input("두번째 정수 입력 >>"))
op = input("연산자 입력(+,-)>> ")
result = cal(num1,num2,op)
print("결과: {}".format(result))
```

```python
def isDuplicate(li1):
    for j in range(len(li1)):
        while j <len(li1):
            for i in range(len(li1)-j):
                if li1[i]==li1[i+1]:
                    return True
            
    return False
    
```

```python
def strikeBallCounting(li1,li2):
    ball =0
    strike =0
    for i in range(len(li1)-1):
        for k in range(len(li1)-1):
            if li1[i]==li2[k]:
                if i==k:
                    strike+=1
                else:
                    ball+=1
    return strike,ball
        
            
```

```python
li1=[1,7,9]
li2 =[1,5.6]
st,ba = strikeBallCounting(li1,li2)
print(st,ba)
```

```python
print("Game Start")
randomList = createRandomNumber(3)
while True:
    inputList = []
    for i in range(len(randomList)):
        num = int(input(f'{i+1}번째 숫자 입력: '))
        inputList.append(num)
    strike, ball = strikeBallCounting(randomList,inputList)
    print(f"strike : {strike}\t ball : {ball}")
    if strike ==len(randomList):
        print("게임종료!!")
        break
    

```
